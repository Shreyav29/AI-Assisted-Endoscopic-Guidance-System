## MODEL in PYTORCH 

```python
import os
import cv2
import pandas as pd
import numpy as np
import random

# PyTorch imports
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# Scikit-learn for splitting and label encoding
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# --------------------------
# 1. Define Paths and Parameters
# --------------------------

# Folder that contains subfolders with images
images_folder = '/home/jayanth/ConsultingProject/DATA/Pooled_Data/Images/'
# Subfolder names (each containing images)
subfolders = ["1_Vocal cords - Copy", "2_stomach photos - Copy", "3_Proper ett depth - Copy", "4_Trachea 1 - Copy"]

# CSV file containing image names and dot coordinates
csv_file = "TrueNorth.csv"

# Define the target image size (square)
IMG_SIZE = 416

# --------------------------
# 2. Helper Function: Get Grid Label from Coordinates
# --------------------------
def get_grid_label(x, y, img_width, img_height):
    """
    Divide an image into a 3x3 grid and determine in which cell (direction) the point (x,y) falls.
    Returns labels like 'north-west', 'center', 'south-east', etc.
    """
    cell_width = img_width / 3
    cell_height = img_height / 3

    # Compute column and row indices; clamp to maximum index of 2
    col = min(int(x // cell_width), 2)
    row = min(int(y // cell_height), 2)

    vertical = ["north", "center", "south"]
    horizontal = ["west", "center", "east"]

    # If the dot is exactly in the center cell, label it "center"
    if row == 1 and col == 1:
        return "center"
    return f"{vertical[row]}-{horizontal[col]}"

# --------------------------
# 3. Load CSV and Prepare the Dataset
# --------------------------
print("Loading images and labels...")
df = pd.read_csv(csv_file)
# Uncomment next line to use a subset for testing:
# df = df.head(20)

images = []    # To store resized images (as numpy arrays)
labels = []    # To store computed grid labels (strings)
filenames = [] # To keep track of filenames (if needed)

# Loop through each row in the CSV file
for idx, row in df.iterrows():
    img_name = row["Image"].strip()  # Remove extra spaces
    x_coord = int(row["X_coord"])     # Original x-coordinate
    y_coord = int(row["Y_coord"])     # Original y-coordinate
    
    found = False
    # Look for the image in each subfolder
    for sub in subfolders:
        img_path = os.path.join(images_folder, sub, img_name)
        if os.path.exists(img_path):
            # Load image using OpenCV
            image = cv2.imread(img_path)
            if image is None:
                print(f"Warning: Could not read image {img_path}.")
                continue

            # Draw a red dot on the image (BGR: (0,0,255))
            cv2.circle(image, (x_coord, y_coord), radius=10, color=(0, 0, 255), thickness=-1)
            
            # Get original dimensions
            orig_height, orig_width = image.shape[:2]
            
            # Resize the image to (IMG_SIZE, IMG_SIZE)
            image_resized = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
            
            # Scale the dot coordinates according to the new image size
            x_scaled = int(x_coord * (IMG_SIZE / orig_width))
            y_scaled = int(y_coord * (IMG_SIZE / orig_height))
            
            # Determine the grid label from the scaled coordinates
            grid_label = get_grid_label(x_scaled, y_scaled, IMG_SIZE, IMG_SIZE)
            
            # Append to lists
            images.append(image_resized)
            labels.append(grid_label)
            filenames.append(img_name)
            
            print(f"Loaded {img_path} with label: {grid_label}")
            found = True
            break  # Stop after finding the image

    if not found:
        print(f"Image {img_name} not found in any folder.")

# Convert list of images to NumPy array and normalize pixel values to [0,1]
images_np = np.array(images, dtype=np.float32) / 255.0
print("Total images loaded:", images_np.shape[0])

print("Final labels in grid : " , list(set(labels)))
```

```python

# --------------------------
# 4. Encode Labels and Create the PyTorch Dataset Class
# --------------------------
le = LabelEncoder()
labels_encoded = le.fit_transform(labels)

class ImageDataset(Dataset):
    def __init__(self, images, labels):
        """
        images: NumPy array of shape (N, H, W, C)
        labels: NumPy array of integer labels
        """
        self.images = images
        self.labels = labels

    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        img = self.images[idx]
        label = self.labels[idx]
        # Convert image from HWC to CHW and then to a torch tensor
        img = torch.from_numpy(img).permute(2, 0, 1)
        return img, label

# --------------------------
# Step A: Split into Train/Test/Holdout Sets and Save Them
# --------------------------
# Split first into 70% train and 30% temporary
X_train, X_temp, y_train, y_temp, fn_train, fn_temp = train_test_split(
    images_np, labels_encoded, filenames, test_size=0.3, random_state=42)

# Split the temporary set equally into 15% test and 15% holdout
X_test, X_holdout, y_test, y_holdout, fn_test, fn_holdout = train_test_split(
    X_temp, y_temp, fn_temp, test_size=0.5, random_state=42)

print("Train set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
print("Holdout set size:", X_holdout.shape[0])

# Create folders inside the TrueNorth folder to save the splits
base_dir = "TrueNorth"
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")
holdout_dir = os.path.join(base_dir, "holdout")
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(holdout_dir, exist_ok=True)

def save_images(dataset, filenames_list, target_dir):
    for img, fn in zip(dataset, filenames_list):
        img_save = (img * 255).astype(np.uint8)
        cv2.imwrite(os.path.join(target_dir, fn), img_save)

save_images(X_train, fn_train, train_dir)
save_images(X_test, fn_test, test_dir)
save_images(X_holdout, fn_holdout, holdout_dir)

print("Datasets saved:")
print("Train:", train_dir)
print("Test:", test_dir)
print("Holdout:", holdout_dir)

# Create PyTorch Dataset objects for each split
train_dataset = ImageDataset(X_train, y_train)
test_dataset = ImageDataset(X_test, y_test)
holdout_dataset = ImageDataset(X_holdout, y_holdout)

# Create DataLoaders for training (and testing)
batch_size = 8  # Adjust based on available memory
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python

# --------------------------
# 5. Define the CNN Model in PyTorch
# --------------------------
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # First convolution: input 3 channels, output 32 channels
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        # Second convolution: 32 to 64 channels
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        # Calculate the feature map size after two poolings (IMG_SIZE/4 x IMG_SIZE/4)
        self.fc1 = nn.Linear(64 * (IMG_SIZE // 4) * (IMG_SIZE // 4), 128)
        self.dropout = nn.Dropout(0.5)
        # Output layer: 9 classes for the 3x3 grid
        self.fc2 = nn.Linear(128, 9)
    
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
model = CNN().to(device)
print(model)

# --------------------------
# 6. Train the Model
# --------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 50 # Adjust as needed

print("Starting training...")
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for batch_idx, (inputs, targets) in enumerate(train_loader):
        inputs, targets = inputs.to(device), targets.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")
```

```python
# --------------------------
# 7. Evaluate on the Test Set
# --------------------------

model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, targets in test_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += targets.size(0)
        correct += (predicted == targets).sum().item()
accuracy = correct / total
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# --------------------------
# Step B: Run Predictions on the Entire Holdout Set and Save Annotated Images
# --------------------------
holdout_predictions_dir = os.path.join(base_dir, "holdoutset_predictions")
os.makedirs(holdout_predictions_dir, exist_ok=True)

# Create a DataLoader for the holdout set (if not already created)
holdout_loader = DataLoader(holdout_dataset, batch_size=batch_size, shuffle=False)

# Loop over the holdout set, predict, annotate, and save images
model.eval()
with torch.no_grad():
    # We loop over indices to retrieve corresponding filenames from fn_holdout list
    for i in range(len(holdout_dataset)):
        img_tensor, _ = holdout_dataset[i]
        img_input = img_tensor.unsqueeze(0).to(device)
        output = model(img_input)
        _, pred_class = torch.max(output.data, 1)
        pred_class = pred_class.item()
        pred_label = le.inverse_transform([pred_class])[0]
        
        # Prepare image for display: convert from CHW to HWC and [0,1] to [0,255]
        img_np = img_tensor.permute(1, 2, 0).cpu().numpy() * 255.0
        img_np = img_np.astype(np.uint8)
        
        # Overlay predicted label on the image using OpenCV
        cv2.putText(img_np, pred_label, (10, 40), cv2.FONT_HERSHEY_SIMPLEX,
                    fontScale=1, color=(255, 255, 255), thickness=2)
        
        # Save annotated image using the corresponding filename from the holdout set
        output_fn = fn_holdout[i]
        cv2.imwrite(os.path.join(holdout_predictions_dir, output_fn), img_np)

print("Holdout set predictions saved in:", holdout_predictions_dir)
```

```python
# --------------------------
# 8. Calculate and Save Metrics, and Plot Performance Graphs
# --------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Collect predictions on the test set (if not already collected)
all_preds = []
all_true = []
model.eval()
with torch.no_grad():
    for inputs, targets in test_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_true.extend(targets.cpu().numpy())

# Calculate overall accuracy
test_acc = accuracy_score(all_true, all_preds)
print("Test Accuracy:", test_acc)

# Generate detailed classification report as a dictionary
report_dict = classification_report(all_true, all_preds, target_names=le.classes_, output_dict=True)

# Convert the report to a DataFrame and save to an Excel file
df_report = pd.DataFrame(report_dict).transpose()
report_path = os.path.join(base_dir, "classification_report.csv")
df_report.to_csv(report_path, index=True)
print(f"Classification report saved to {report_path}")

# Compute confusion matrix and save as Excel
conf_matrix = confusion_matrix(all_true, all_preds)
df_conf_matrix = pd.DataFrame(conf_matrix, index=le.classes_, columns=le.classes_)
conf_matrix_path = os.path.join(base_dir, "confusion_matrix.csv")
df_conf_matrix.to_csv(conf_matrix_path, index=True)
print(f"Confusion matrix saved to {conf_matrix_path}")

# Plot confusion matrix heatmap and save as an image
plt.figure(figsize=(8, 6))
sns.heatmap(df_conf_matrix, annot=True, cmap="Blues", fmt="d")
plt.title("Confusion Matrix")
plt.ylabel("Actual")
plt.xlabel("Predicted")
conf_matrix_plot_path = os.path.join(base_dir, "confusion_matrix.png")
plt.savefig(conf_matrix_plot_path)
plt.close()
print(f"Confusion matrix plot saved to {conf_matrix_plot_path}")

# If you recorded training losses in a list called 'train_losses' during training, plot them:
if 'train_losses' in globals():
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, num_epochs + 1), train_losses, marker='o')
    plt.title("Training Loss per Epoch")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    training_loss_plot_path = os.path.join(base_dir, "training_loss.png")
    plt.savefig(training_loss_plot_path)
    plt.close()
    print(f"Training loss plot saved to {training_loss_plot_path}")


# --------------------------
# 8. Save the Model as a .pt File
# --------------------------
model_save_path = os.path.join(base_dir, "model.pt")
torch.save(model.state_dict(), model_save_path)
print(f"Model saved to {model_save_path}")


# --------------------------
# 8.1. Save the Model as a .onnx File
# --------------------------

# Reload model to CPU and prepare for ONNX export
model_for_onnx = CNN()
model_for_onnx.load_state_dict(torch.load(model_save_path, map_location="cpu"))
model_for_onnx.eval()

# Create dummy input matching your model input
dummy_input = torch.randn(1, 3, 416, 416)

# Define ONNX export path
onnx_path = os.path.join(base_dir, "TrueNorth_model.onnx")

# Export to ONNX
torch.onnx.export(
    model_for_onnx,
    dummy_input,
    onnx_path,
    input_names=["input"],
    output_names=["output"],
    dynamic_axes={"input": {0: "batch_size"}, "output": {0: "batch_size"}},
    export_params=True,
    opset_version=12,
    do_constant_folding=True
)

print(f" ONNX model exported and saved as {onnx_path}")







# --------------------------
# 9. Convert to Mobile Optimized Version 
# --------------------------
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.mobile_optimizer import optimize_for_mobile


# Define your CNN architecture (adjust as needed)
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Example architecture; modify according to your training code
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        # After two poolings, image size becomes (416/4 x 416/4)
        self.fc1 = nn.Linear(64 * (416 // 4) * (416 // 4), 128)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 9)  # Assuming 9 classes for the grid

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Step 1: Reinitialize your model and load the state dictionary from "model.pt"
model = CNN()
state_dict = torch.load(model_save_path, map_location="cpu")  # "model.pt" is the file you saved earlier
model.load_state_dict(state_dict)
model.eval()  # Set the model to evaluation mode

# Step 2: Create a dummy input that matches your model's expected input size (batch=1, 3 channels, 416x416)
dummy_input = torch.rand(1, 3, 416, 416)

# Step 3: Convert the model to TorchScript using tracing
traced_model = torch.jit.trace(model, dummy_input)
torchscript_path =  os.path.join(base_dir, "TrueNorth_model_torchscript.pt")
traced_model.save(torchscript_path)
print(f" TorchScript model saved as {torchscript_path}")

# Step 4: Optimize the TorchScript model for mobile deployment
optimized_model = optimize_for_mobile(traced_model)
optimized_model_path =  os.path.join(base_dir, "TrueNorth_model_mobile_optimized.pt")
optimized_model.save(optimized_model_path)
print(f" Optimized TorchScript model saved for mobile as {optimized_model_path}")

```
